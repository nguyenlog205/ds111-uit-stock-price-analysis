```python
from vnstock import Quote
from vnstock import Listing
from vnstock import Finance
```

## 1. General use

### 1.1. List available Stock ID

```python
listing = Listing(source='VCI')
```

```python
listing.all_symbols()
```

```python
listing.industries_icb()
```

## 2. Crawl historical stock price dataset

```python
quote = Quote(symbol='FPT', source='VCI')
```

```python
historical_price = quote.history(start='2007-01-01', end='2025-10-09')
```

```python
historical_price.to_csv('../data/fpt_historical_price.csv', index=False)
```

## 3. Crawl historical Finance Report dataset

```python
finance = Finance(symbol='FPT', source='VCI')
```

```python
income_statement_df = finance.income_statement(period='quarter', lang='vi')
```

```python
income_statement_df
```

```python
income_statement_df.to_csv('../data/fpt_income_statement.csv', index=False)
```
